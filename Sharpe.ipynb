```python
MC = 100000
vset = []
vset2 = []
for m in np.arange(MC):
    Z = np.random.choice(X,len(X))
    vset.append(Z.mean()) 
    vset2.append(Z.var())
vset = np.array(vset)  
vset2 = np.array(vset2)
```

```python
vset
```

```python
vm = vset.mean()
vm2 = vset2.mean()
```

```python
SE = np.sqrt(np.dot(vset-vm,vset-vm)/(MC-1))
SE2 = np.sqrt(np.dot(vset2-vm2,vset2-vm2)/(MC-1))
```
